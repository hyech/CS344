Due to the lack of pictures of the two buildings I selected (KE Apartments Gamma and Theta) in our class dataset, I have opted to include an "Other" category in order to complete this homework. I placed twelve random images from Adam DenHaan's (add22) dataset into the training dataset and reserved the rest (58 photos) for the test dataset.

```python
# Create path object
path = Path('/home/hj48/344/hw03/train')
```

```python
path
```

```python
image_files = get_image_files(path).sorted()
image_files[0].name
```

```python
# Creating a smaller dataset
small_image_files = [img for img in image_files if int(img.name[-6]) < 1]
```

```python
# Seed 12345
set_seed(12345, reproducible=True)
```

```python
# Seed 54321
set_seed(54321, reproducible=True)
```

```python
# Seed 13542
set_seed(13542, reproducible=True)
```

```python
# Construct the data loader
dataloaders = ImageDataLoaders.from_path_func(
    path=path,
    fnames=image_files,
    # Use a 20% validation split, with a seed of 42 for reproducibiliy
    valid_pct=0.2,
    seed=42,
    label_func=parent_label,
    # Resize all images to 224x224 pixels
    item_tfms=RandomResizedCrop(224, min_scale=0.7),
    # Set batch size to 2 b/c we have a small dataset
    bs=2,
)
```

```python
dataloaders.train.show_batch()
```

```python
learn = cnn_learner(
    dls=dataloaders,
    arch=resnet34,
    metrics=error_rate
)
learn.fine_tune(epochs=1)
```

```python
clf_error_rates = [0.071429, 0.214286, 0.214286]
```

```python
clf_small_error_rates = [0.000000, 0.333333]
```

```python
# Display confusion matrix
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
# Show predictions/actual/loss/probability
interp.plot_top_losses(5, nrows=1)
```

## HW 03 Code

```python
# Create a test path and get the images
test_path = Path('/home/hj48/344/hw03/test')
test_image_files = get_image_files(test_path)
```

```python
test_image_files[0].name
```

```python
# Create a test dataloader
test_dl = dataloaders.test_dl(test_image_files, with_labels=True)
```

```python
# Get interpretations
interp = ClassificationInterpretation.from_learner(learn, dl=test_dl)
```

```python
# Get accuracy scores and predictions
interp.plot_top_losses(k = 5)
interp.print_classification_report()
```

```python
# Long-form code to get accuracy score
(
    # Make a Tensor of Trues and Falses, True if the classifier got the corresponding image right
    (
        # the predictions that the model made on the test set
        interp.decoded
        # compare with the target labels provided by the DataLoader
        == interp.targs
    )
    # convert True to 1.0, False to 0.0
    .to(float)
    # Compute the fraction of True's.
    .mean()
)
```

```python
# Short-form version
accuracy(interp.preds, interp.targs)
```
